```python
%matplotlib inline
import numpy as np
import pandas as pd
import scipy
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#data import
yelp = pd.read_csv('yelp_labelled.txt', sep="\t", header=None)

yelp.columns = ['message','sentiment']
#print(yelp)
```

# Classifier 1 - a "great" baseline

```python
#data import
yelp = pd.read_csv('yelp_labelled.txt', sep="\t", header=None)

yelp.columns = ['message','sentiment']
#print(yelp)

pos_keywords = ['great']

for key in pos_keywords:
    # Note that we add spaces around the key so that we're getting the word,
    # not just pattern matching.
    yelp[str(key)] = yelp.message.str.contains(
        ' ' + str(key) + ' ',
        case=False
    )

plt.figure(figsize=(8,6))
sns.heatmap(yelp.corr())
```

```python
# define data and target
data = yelp[pos_keywords]
target = yelp['sentiment']

# Our data is binary / boolean, so we're importing the Bernoulli classifier.
from sklearn.naive_bayes import BernoulliNB

# Instantiate our model and store it in a new variable.
bnb = BernoulliNB()

# Fit our model to the data.
bnb.fit(data, target)

# Classify, storing the result in a new variable.
y_pred = bnb.predict(data)

# Display our results.
print("Number of mislabeled points out of a total {} points : {}".format(
    data.shape[0],
    (target != y_pred).sum()
))
accuracy = (target == y_pred).sum()/(data.shape[0])*100
print("Accuracy of this model: {}%".format(round(accuracy,2)))

from sklearn.metrics import confusion_matrix
conf = confusion_matrix(target, y_pred)
print('True positive:{}'.format(conf[0][0]))
print('False positive:{}'.format(conf[0][1]))
print('True negative:{}'.format(conf[1][0]))
print('False negative:{}'.format(conf[1][1]))
```

```python
acc = bnb.score(data,target)
print(acc)
```
