```python
# Drill answer code:


df_is=pd.DataFrame({"x":list(range(1,11))*10,
                "y":list(range(1,21,2))*10,
                "id":np.repeat(range(10),10),
                "time":list(range(10))*10})

for i in range(10):
    df_is.loc[df_is['id']==i,"y"]=df_is.loc[df_is['id']==i,"y"]+i*3
    df_is.loc[df_is['id']==i,"y"]=df_is.loc[df_is['id']==i,"y"]*i

# Plotting the data: Same slope, different intercepts.
sns.lmplot(x="x", y="y", hue="id", data=df_is)
plt.show()
```
