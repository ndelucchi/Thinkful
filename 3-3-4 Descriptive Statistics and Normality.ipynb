```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
binomial = np.random.binomial(n=10, p=.5, size=100)
plt.hist(binomial)
'''Adding mean and standard deviation dash lines'''
plt.axvline(binomial.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(binomial.mean() + binomial.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(binomial.mean() - binomial.std(), color='c', linestyle='dashed', linewidth=2)
plt.show()
"""In thise case, the mean and standard deviation lines provide a useful description"""
```

```python
exponential = np.random.exponential(scale=7.5, size=100)
plt.hist(exponential)
'''Adding mean and standard deviation dash lines'''
plt.axvline(exponential.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(exponential.mean() + exponential.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(exponential.mean() - exponential.std(), color='c', linestyle='dashed', linewidth=2)
plt.show()
"""In thise case, the mean and standard deviation lines doesn't provide a useful description"""
```

```python
gamma = np.random.gamma(shape=4.3, scale=2.25, size=100)
plt.hist(gamma)
'''Adding mean and standard deviation dash lines'''
plt.axvline(gamma.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(gamma.mean() + gamma.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(gamma.mean() - gamma.std(), color='c', linestyle='dashed', linewidth=2)
plt.show()
"""In thise case, the mean and standard deviation lines provide a useful description"""
```

```python
lognormal = np.random.lognormal(mean=38.03,sigma=5.8, size=100)
plt.hist(lognormal)
'''Adding mean and standard deviation dash lines'''
plt.axvline(lognormal.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(lognormal.mean() + lognormal.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(lognormal.mean() - lognormal.std(), color='c', linestyle='dashed', linewidth=2)
plt.show()
"""In thise case, the mean and standard deviation lines provide a useful description"""
```

```python
pareto = np.random.pareto(a=3.8, size=100)
plt.hist(pareto)
'''Adding mean and standard deviation dash lines'''
plt.axvline(pareto.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(pareto.mean() + pareto.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(pareto.mean() - pareto.std(), color='c', linestyle='dashed', linewidth=2)
plt.show()
"""In thise case, the mean and standard deviation line don't provide a useful description"""
```

```python
uniform = np.random.uniform(low=6,high=18, size=100)
plt.hist(uniform)
'''Adding mean and standard deviation dash lines'''
plt.axvline(uniform.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(uniform.mean() + uniform.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(uniform.mean() - uniform.std(), color='c', linestyle='dashed', linewidth=2)
plt.show()
"""In thise case, the mean and standard deviation lines don't provide a useful description"""
```

```python
"""The mean and standard deviation lines only provide a useful description when the distribution isn't skewed or uniform or have extreme outliers"""
```

```python
rand1 = np.random.normal(loc=5, scale=0.5, size=1000)
rand2 = np.random.normal(loc=10, scale=1, size=1000)
rand3 = rand1 + rand2
plt.hist(rand3)
'''Adding mean and standard deviation dash lines'''
plt.axvline(rand3.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(rand3.mean() + rand3.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(rand3.mean() - rand3.std(), color='c', linestyle='dashed', linewidth=2)
plt.show()
```

```python
"""As a near-normal distribution, the mean and standard deviation lines provide a meaningful, useful description of the variable."""
```
