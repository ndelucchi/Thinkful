```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import ttest_ind
%matplotlib inline
```

```python
print("Initial set up")
pop1 = np.random.binomial(n=10, p=0.2, size=10000)
pop2 = np.random.binomial(n=10, p=0.5, size=10000)
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)
print("Sample1 mean is",sample1.mean())
print("Sample2 mean is",sample2.mean())
print("Sample1 standard deviation is",sample1.std())
print("Sample2 standard deviation is",sample2.std())
plt.figure()
plt.hist(sample1,color="c", alpha=0.5, label="Sample1")
plt.hist(sample2,color="b", alpha=0.5, label="Sample2")
plt.legend(loc='upper right')
plt.show()
```
